```python
# Set up matplotlib
import matplotlib.pyplot as plt
#magic function that makes it not dump to another thing and just prints here (how neat)
%matplotlib inline 
from astropy.cosmology import FlatLambdaCDM
import astropy.units as u
cosmo = FlatLambdaCDM(H0=70*u.km/u.s/u.Mpc, Om0=0.3)
#this sets the hubble constant = 70km/s/Mpc since FlatLambdaCDM doens't assume it
#we need something to plot vs redshift so we will use angular diameter distance
import numpy as np
zvals = np.arange(0,6,.1)
dist = cosmo.angular_diameter_distance(zvals)
#the above 3 lines will generate an array of values 0-6 in intervals of .1 and give an angular distance for each value
#the zvals in this case are values of redshift
#we will now plot the figure as angular size vs redshift
fig = plt.figure(figsize = (6,4))
#ax = fig.add_subplot(111)  #add a subplot in the first quadrant with 1x1 scale
#ax.plot(zvals,dist)


#Now we are going to tweak the graph a little and make it more refined
ages = np.array([13, 10, 8, 6, 5, 4, 3, 2, 1.5, 1.2, 1])*u.Gyr #simply generates an array of universe ages
from astropy.cosmology import z_at_value #this will allow you to look at z with different specified values
ageticks = [z_at_value(cosmo.age, age) for age in ages] #this allows to input different ages and gives the z value for the array "ages"
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)
ax.plot(zvals,dist)
ax2 = ax.twiny() #creates a twin axis that shares the y axis
ax2.set_xticks(ageticks) #sets the values of the ax2 axis
ax2.set_xticklabels(['{:g}'.format(age) for age in ages.value]) #formats age to be in :g form (float)(?)
zmin, zmax = 0.0 , 5.9
ax.set_xlim(zmin,zmax)
ax2.set_xlim(zmin,zmax)
ax2.set_xlabel('Time since Big Bang (Gyr)')
ax.set_xlabel('Redshift')
ax.set_ylabel('Angular diameter distance (Mpc)')
ax.minorticks_on()

from astropy.cosmology import Planck13
dist2 = Planck13.angular_diameter_distance(zvals)

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)
ax.plot(zvals, dist2, label='Planck 2013')                  #making legend
ax.plot(zvals, dist, label='$h=.7,\ \Omega_\Lambda=.7$')    #making legend
ax.legend(frameon=0, loc='lower right')                     #making legend
ax2 = ax.twiny()
ax2.set_xticks(ageticks)
ax2.set_xticklabels(['{:g}'.format(age) for age in ages.value])
zmin, zmax = 0.0, 5.9
ax.set_xlim(zmin,zmax)
ax2.set_xlim(zmin,zmax)
ax2.set_xlabel('Time since Big Bang (Gyr)')
ax.set_xlabel('Redshift')
ax.set_ylabel('Angular diameter distance (Mpc)')
ax.minorticks_on()
ax.set_ylim(0,1890)

#I wanted to see how WMAP7 compared since it is much older
from astropy.cosmology import WMAP7
dist3 = WMAP7.angular_diameter_distance(zvals)

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)
ax.plot(zvals, dist3, label='WMAP7')
ax.plot(zvals, dist2, label='Planck 2013')                  #making legend
ax.plot(zvals, dist, label='$h=.7,\ \Omega_\Lambda=.7$')    #making legend
ax.legend(frameon=0, loc='lower right')                     #making legend
ax2 = ax.twiny()
ax2.set_xticks(ageticks)
ax2.set_xticklabels(['{:g}'.format(age) for age in ages.value])
zmin, zmax = 0.0, 5.9
ax.set_xlim(zmin,zmax)
ax2.set_xlim(zmin,zmax)
ax2.set_xlabel('Time since Big Bang (Gyr)')
ax.set_xlabel('Redshift')
ax.set_ylabel('Angular diameter distance (Mpc)')
ax.minorticks_on()
ax.set_ylim(0,1890)
```
